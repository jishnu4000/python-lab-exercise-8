```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./heart_failure_clinical_records_dataset.csv')

df.head()
```

```python
plt.hist(df['age'], bins=20, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['age'], df['ejection_fraction'])
plt.title('Age vs Ejection Fraction')
plt.xlabel('Age')
plt.ylabel('Ejection Fraction')
plt.show()
```

```python
plt.plot(df['time'], df['serum_creatinine'])
plt.title('Serum Creatinine Levels Over Time')
plt.xlabel('Time')
plt.ylabel('Serum Creatinine')
plt.show()
```

```python
df['smoking'].value_counts().plot(kind='bar')
plt.title('Smoking Status Bar Plot')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(df['DEATH_EVENT'].value_counts(), labels=['Alive', 'Dead'], autopct='%1.2f%%')
plt.title('Proportion of Death Events')
plt.show()
```

```python
plt.fill_between(df['age'], df['sex'].cumsum(), label='Male')
plt.fill_between(df['age'], df['sex'].cumsum(), df.shape[0], label='Female')
plt.title('Cumulative Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Cumulative Count')
plt.legend()
plt.show()
```

```python
sns.boxplot(x='DEATH_EVENT', y='age', data=df)
plt.title('Age Distribution by Death Event')
plt.xlabel('Death Event')
plt.ylabel('Age')
plt.show()
```

```python
sns.pairplot(df[['age', 'creatinine_phosphokinase', 'ejection_fraction', 'serum_creatinine', 'DEATH_EVENT']], hue='DEATH_EVENT')
plt.show()
```
